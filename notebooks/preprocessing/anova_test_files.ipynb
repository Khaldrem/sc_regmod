```python
def detect_pvalue(pvalue):
    if pvalue <= 0.05:
        return True
    return False




"""
    Por cada archivo:
        1. Por cada columna del archivo:
            2. Por cada fenotipo:
                2.1 Extraer desde el csv fenotipos los valores en base a los ids
                    y agrupar valores por cada base en la columna. Ej:

                    "SM300-Efficiency": {
                        "t": [0.01, 0.02, 0.03 ...],
                        "g": [0.04, 0.05, 0.06 ...]
                    },
                    "SM300-Rate": {
                        "t": [0.12, 0.21, 0.43 ...],
                        "g": [0.99, 0.77, 0.32 ...]
                    },
                    ...
            
            3. Evaluar anova por cada fenotipo
                3.1 Obtener el pvalue de la columna
                3.2 Detectar si ese pvalue <= 0.05
                    3.2.1 Si se detecta positivo se a√±ade a la lista de los datos que quedaran
                    3.2.2 Si se detecta positivo se aumenta el contador de fenotipos que influyeron en el archivo. Ej.

                        {
                            "SM300-Efficiency": 1,
                            "SM300-Rate": 0,
                            ...
                        }

                        Esto permite conocer que fenotipo fue el que mas aporto al archivo

            4. Obtener la cantidad de bases por archivo. Ej.

                {
                    "t": 740,
                    "a": 89,
                    "g": 10,
                    ...
                }
            

                
"""


#Por cada archivo!

#Por cada columna
for 


bases_in_file = {}
anova_dataset = []

#Por cada columna en los datos
for col in range(test_data.get_alignment_length()):

    #Obtengo los valores unicos de la fila
    unique_keys = set(list(test_data[:, col]))
    if len(unique_keys) != 1:
        anova_data = {}
        for key in unique_keys:
            anova_data[key] = []

        #anova_data ahora tiene arreglos vacios en base a las llaves unicas de la columna

        #Por cada fila en el archivo
        for row in range(len(test_data)):
            #Obtengo la base
            base = test_data[row, col]

            #Si el id se encuentra entre las llaves de los fenotipos (asegura que si exista el dato)
            if test_data[row].id in  pheno.loc[:, "Standard"].tolist():
                #Se extrae el dato desde los fenotipos con igual ID y en base a la columna (ej. "SM300-Efficiency" o "SM300-Rate")
                pheno_data = pheno[pheno["Standard"] == test_data[row].id]["SM300-Efficiency"].astype('float64').values[0]
                #Se a√±ade el valor
                anova_data[base].append(pheno_data)
        
        anova_dataset.append(anova_data)
    else:
        print(f"col: {col} has just 1 element")


#va con un -1 en vez de un 3
phenotypes_keys = pheno.keys()[2:3]

#Poseemos todos los datos por columna para realizar el anova, por cada columna
for data in anova_dataset:
    has_array_0 = False

    for key in data:
        if len(data[key]) == 0:
            has_array_0 = True
    
    if not has_array_0:
        #Por cada tipo de fenotipo
        for phenotype in phenotypes_keys:
            anova_test = {}
            anova_test[phenotype] = data
            df_to_test = pd.DataFrame.from_dict(anova_test, orient='index')
            results = stats.f_oneway(*df_to_test.loc[phenotype])
            print(detect_pvalue(results.pvalue))
            
    #else:
    #    print(data)

    #print("===============")

#print(anova_dataset)
    
```
