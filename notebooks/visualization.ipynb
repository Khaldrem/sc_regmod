```python
import matplotlib

chromosome = {
    "haploide-euploide": phenotype_df[phenotype_df["Haploide-Diploide"] == "haploide-euploide"].count()[0],
    "diploides-euploides": phenotype_df[phenotype_df["Haploide-Diploide"] == "diploides-euploides"].count()[0],
    "---": phenotype_df[phenotype_df["Haploide-Diploide"] == "---"].count()[0]
}

perc =  [float(i) for i in chromosome.values()]
sns.set(rc={'figure.figsize':(12,9)})
b = sns.barplot(x=list(chromosome.keys()),y=perc)
patches = [matplotlib.patches.Patch(color=sns.color_palette()[i], label=t) for i,t in enumerate(t.get_text() for t in b.get_xticklabels())]
plt.legend(handles=patches, loc="upper right") 
```

```python
ecological = {}

set_eco = set(phenotype_df["Ecological info"])

for key in set_eco:
    ecological[key] = phenotype_df[phenotype_df["Ecological info"] == key].count()[0]

perc =  [float(i) for i in ecological.values()]
sns.set(rc={'figure.figsize':(12,9)})
b = sns.barplot(x=list(ecological.keys()),y=perc)
patches = [matplotlib.patches.Patch(color=sns.color_palette()[i], label=t) for i,t in enumerate(t.get_text() for t in b.get_xticklabels())]
plt.legend(handles=patches, loc="upper left") 
```
