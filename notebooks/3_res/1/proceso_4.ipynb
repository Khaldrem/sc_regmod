```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_pred(data, phenotype, exp_num, save_path):
    plt.figure(figsize=(15, 10))
    plt.scatter(data["y_test"], data["Ridge_y_pred"], marker='.', color="blue", label="Ridge")
    plt.scatter(data["y_test"], data["Lasso_y_pred"], marker='*', color="green", label="Lasso")
    plt.scatter(data["y_test"], data["RandomForestRegressor_y_pred"], marker='>', color="red", label="Random Forest")
    plt.legend(loc="upper right")
    plt.xlabel("Verdadero", fontsize=18)
    plt.ylabel("Predecido", fontsize=18)
    plt.title(f"{phenotype}", fontsize=20)
    plt.savefig(f"{save_path}/{phenotype}_pred.png")
```

```python
phenotypes = [
    "SM300-Efficiency", "SM300-Rate", "SM300-Lag", "SM300-AUC",
    "SM60-Efficiency",  "SM60-Rate",  "SM60-Lag",  "SM60-AUC",
    "Ratio-Efficiency", "Ratio-Rate", "Ratio-Lag", "Ratio-AUC",
]


EXP_NUM = 1
SAVE_PATH = "/home/khaldrem/Desktop/tesis/imagenes/4/1/3"
```

```python
for pt in phenotypes:
    data = pd.read_csv(f"/home/khaldrem/code/sc_regmod/dataset/models/final/{EXP_NUM}/pred/{pt}_y_pred.csv")
    plot_pred(data, pt, EXP_NUM, SAVE_PATH)
```
